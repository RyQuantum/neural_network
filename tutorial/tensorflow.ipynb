```python
from datetime import datetime
```

```python
import tensorflow as tf
import pandas as pd
```

```python
data = pd.read_csv('./dataset/Income1.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.scatter(data.Education, data.Income)
```

```python
x = data.Education
y = data.Income
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
                 loss='mse'
)
```

```python
a = datetime.now()
history = model.fit(x, y, epochs=9000)
b = datetime.now()
(b - a).seconds
```

```python
model.predict(x)
```
